```python
import multiprocessing
import subprocess

verbose = True 
capture_output = False if verbose else True 
```

### Launch the server

```python
script = "run_server.py"

def _run_script(script, capture_output):
    subprocess.run(["python3", script], capture_output=capture_output)

# Create a background process
process = multiprocessing.Process(target=_run_script, args=(script, capture_output))

process.start()
print(f"{script} is running in the background...")
```

### Launch the interface

```python
script = "run_interface.py"

# Change the code to use panel serve instead of python3 
def _run_script(script, capture_output):
    subprocess.run(["panel", "serve", script], capture_output=capture_output)
```

```python
# Create a background process
process = multiprocessing.Process(target=_run_script, args=(script, capture_output))

process.start()
print(f"{script} is running in the background...")
```

### Control the interface from the notebook


```python
import numpy as np

from vivarium.controllers.notebook_controller import NotebookController

controller = NotebookController()
```

```python
controller.simulator_config
```

```python
controller.agents[0].diameter = 10.
```

```python
# If you launch another cell, it will keep printing the result of the subprocess below the cell (but we can use verbose=False)
controller.agents[2].color = 'green'
```

```python
def aggression(agent):
    return agent.right_prox, agent.left_prox

def fear(agent):
    return agent.left_prox, agent.right_prox

def love(agent):
    return 1. - agent.left_prox, 1. - agent.right_prox

def shy(agent):
    return 1. - agent.right_prox, 1. - agent.left_prox
```

```python
for ag in controller.agents[:5]:
    ag.color = 'blue'
    ag.detach_all_behaviors()
    ag.attach_behavior(shy)
for ag in controller.agents[5:]:
    ag.color = 'red'
    ag.detach_all_behaviors()
    ag.attach_behavior(aggression)
```
