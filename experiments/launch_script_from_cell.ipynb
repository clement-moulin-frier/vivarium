```python
import multiprocessing
import subprocess

def run_script(script_name='script', verbose=True):
    script = f"{script_name}.py"
    capture_output = False if verbose else True 

    def _run_script(script, capture_output):
        subprocess.run(["python3", script], capture_output=capture_output)

    # Create a background process
    process = multiprocessing.Process(target=_run_script, args=(script, capture_output))

    process.start()
    print(f"{script} is running in the background...")
```

### Test with verbose = True 

```python
verbose = True 

run_script(verbose=verbose)
```

```python
# If you launch another cell, it will keep printing the result of the subprocess below the cell (but we can use verbose=False)
print(f"Useless cell to run while starting the subprocess : {1 + 4 = }")
```

### Test with verbose = False 

```python
verbose = False 

run_script(verbose=verbose)
```
